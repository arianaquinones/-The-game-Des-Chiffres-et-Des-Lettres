```python
import random
import requests
# I helped myself on internet to do this assignement 

# NUMBER GAME FUNCTIONS 

# Generate a target number between 100 and 999
def target_number(a, b):
    return random.randint(a, b)

# Generate six random numbers
def random_numbers(a, b):
    first_digit = random.randint(a, b)
    second_digit = random.randint(a, b)
    third_digit = random.randint(a, b)
    fourth_digit = random.randint(a, b)
    fifth_digit = random.randint(a, b)
    sixth_digit = random.choice([25, 50, 75, 100])
    return first_digit, second_digit, third_digit, fourth_digit, fifth_digit, sixth_digit

# addition
def addition(x, y):
      return x + y

# subtraction
def subtraction(x, y):
    return x - y

# multiplication
def multiplication(x, y):
    return x * y

# division
def division(x, y):
    # Check for division by zero
    if y == 0:
        return "Error: division by zero"
    return x / y

# Evaluate an operation between two numbers
def evaluate_simple_expression(num1, operator, num2):
    if operator == '+':
        return addition(num1, num2)
    elif operator == '-':
        return subtraction(num1, num2)
    elif operator == '*':
        return multiplication(num1, num2)
    elif operator == '/':
        return division(num1, num2)
    else:
        return "Invalid operator"

# Let user decide to add more operations or stop
def continue_operations(result):
    while True:
        user_choice = input("Do you want to continue with more operations? (yes/no): ").lower()
        if user_choice == "yes":
            operator = input("Enter the next operator (+, -, *, /): ")
            num = int(input("Enter the next number: "))
            result = evaluate_simple_expression(result, operator, num)
            print(f"New result: {result}")
        elif user_choice == "no":
            return result
        else:
            print("Invalid choice, please enter 'yes' or 'no'.")

# Calculate the difference between the target and the result
def calculate_difference(target, result):
    return abs(target - result)
# here absolute value is needed in case the result is greater than the target to not find a negative number 

# Calculate the score based on the difference based on the course formula
def calculate_score(target, result):
    difference = calculate_difference(target, result)
    return max(100 - (difference // 10), 0)
# here the function max is used to don t have negative scores 

# ---------- WORD GAME FUNCTIONS ----------

# Define the vowels and consonants
vowels = 'AEIOU'
consonants = 'BCDFGHJKLMNPQRSTVWXYZ'

# Fetch the longest word from the dictionary based on available letters
def fetch_longest_word(letters, word_list):
    longest_word = ""
    letters = sorted(letters)

    # Go through all words and find the longest word that can be created using the provided letters
    for word in word_list:
        word = word.upper()
        word_letters = sorted(word)
        
        # Check if the word can be formed from the provided letters
        if all(word_letters.count(char) <= letters.count(char) for char in word_letters):
            if len(word) > len(longest_word):
                longest_word = word

    # If no valid word is found, return a default message
    return longest_word if longest_word else "No valid word found"

# Calculate the score using the formula S = 100 - (L - l) * 10
def calculate_word_score(player_word, longest_word):
    L = len(longest_word)
    l = len(player_word)
    return max(min(100 - (L - l) * 10, 100), 0)  # Cap the score at 100, and prevent negative scores

# Function to get the letters from the user
def generate_letters():
    letters = []

    for _ in range(10):
        choice = input("Enter 'vowel' or 'consonant': ").strip().lower()

        if choice == 'vowel':
            letters.append(random.choice(vowels))
        elif choice == 'consonant':
            letters.append(random.choice(consonants))
        else:
            print("Invalid input. Please enter either 'vowel' or 'consonant'.")
            continue  # Ask again if invalid input

    return letters

# Load the list of words from the URL, I helped myself on internet 
def load_word_list():
    url = "https://raw.githubusercontent.com/dwyl/english-words/refs/heads/master/words.txt"
    response = requests.get(url)
    if response.status_code == 200:
        word_list = response.text.splitlines()  # Split the text into a list of words
        print(f"Loaded {len(word_list)} words from the dictionary.")
        return word_list
    else:
        print("Failed to load the word list from the URL.")
        return []  # Return an empty list if the URL fails

# initialization of the games 
def play_number_game():
    print("Welcome to the Number Game!")
    target = target_number(100, 999)
    numbers = random_numbers(1, 10)
    print(f"Target number: {target}")
    print(f"Available numbers: {numbers}")
    
    result = int(input("Start with a number from the available ones: "))
    result = continue_operations(result)
    
    score = calculate_score(target, result)
    print(f"Your result: {result}")
    print(f"Target was: {target}")
    print(f"Your score: {score}")


def play_word_game():
    print("Welcome to the Word Game!")
    letters = generate_letters()
    print(f"Your letters are: {''.join(letters)}")
    
    word_list = load_word_list()
    if not word_list:
        print("Unable to play the game due to missing word list.")
        return
    
    player_word = input("Enter your word: ").upper()
    
    longest_word = fetch_longest_word(letters, word_list)
    print(f"The longest possible word is: {longest_word}")
    
    score = calculate_word_score(player_word, longest_word)
    print(f"Your word: {player_word}")
    print(f"Your score: {score}")

# Run the games
if __name__ == "__main__":
    print("Number Game")
    play_number_game()

    print("\n Word Game")
    play_word_game()
```
